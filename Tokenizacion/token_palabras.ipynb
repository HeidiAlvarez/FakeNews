```python
import re
from collections import Counter
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
from nltk.corpus import stopwords
from nltk.tokenize import word_tokenize
from collections import Counter
import nltk
nltk.download('punkt')
nltk.download('stopwords')

import nltk
nltk.download()
nltk.download('punkt')
```

```python
csv_file_path = 'C:\\Users\\juans\\Desktop\\Analisis de datos\\Proyecto Fake News\\FakeNews\\Tokenizacion\\noticias_limpias.csv'
noticias = pd.read_csv(csv_file_path)
```

```python

# Descargar recursos (solo una vez)
nltk.download('stopwords')
nltk.download('punkt')

# Obtener stopwords en espaol
spanish_stopwords = set(stopwords.words('spanish'))

# Tokenizar y filtrar en una pasada
all_tokens = []
for noticia in noticias['body']:
    tokens = word_tokenize(noticia.lower())
    tokens_limpios = [token for token in tokens if token not in spanish_stopwords and token.isalpha()]
    all_tokens.extend(tokens_limpios)

# Obtener los 20 ms comunes
token_counts = Counter(all_tokens)
most_common_tokens = token_counts.most_common(20)
tokens, counts = zip(*most_common_tokens)

plt.figure(figsize=(12, 6))
sns.barplot(x=list(tokens), y=list(counts), palette='viridis')
plt.title('Top 20 Palabras Ms Comunes en Noticias (sin stop words)')
plt.xlabel('Palabras')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45)
plt.show()

```

```python

# Listas de palabras positivas y negativas 
positive_words = {'bueno', 'excelente', 'positivo', 'feliz', 'maravilloso', 'genial', 'fantstico'}
negative_words = {'malo', 'terrible', 'negativo', 'triste',
                    'horrible', 'psimo', 'desastroso'}
# Clasificar palabras
word_sentiments = {}
for token in all_tokens:
    if token in positive_words:
        word_sentiments[token] = 'positiva'
    elif token in negative_words:
        word_sentiments[token] = 'negativa'
    else:
        word_sentiments[token] = 'neutra'
# Contar frecuencias por categora
sentiment_counts = {'positiva': 0, 'negativa': 0, 'neutra': 0}
for token in all_tokens:
    sentiment = word_sentiments[token]
    sentiment_counts[sentiment] += 1
# Visualizar resultados
plt.figure(figsize=(8, 6))
sns.barplot(x=list(sentiment_counts.keys()), y=list(sentiment_counts.values()), palette='pastel')
plt.title('Frecuencia de Palabras por Sentimiento')
plt.xlabel('Sentimiento')
plt.ylabel('Frecuencia')
plt.show()

```
